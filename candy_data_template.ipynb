# cleaning

### 4- drop all the useless columns and  rows with too much NaN , 
> tip : useless columns can be like `Internal ID`and can be also columns with too much NaN 
- rename columns for easy access

```python
candy.drop(['Internal ID', 'Q12: MEDIA [Yahoo]', 'Q12: MEDIA [ESPN]', 'Q12: MEDIA [Daily Dish]', 'Unnamed: 113', 'Q9: OTHER COMMENTS','Q12: MEDIA [Science]','Q8: DESPAIR OTHER', 'Q7: JOY OTHER' ], axis =1, inplace= True)
candy.rename(columns={'Q4: COUNTRY' : 'country', 'Q5: STATE, PROVINCE, COUNTY, ETC' : 'area', 'Q10: DRESS' :'dress', 'Q11: DAY' : 'day', 'Q3: AGE' : 'age'}, inplace= True)
candy.head()
```

### 5 - clean `going out` column 
> tip : you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
candy['Q1: GOING OUT?'] = candy['Q1: GOING OUT?'].fillna('I do not know')
candy['Q1: GOING OUT?'].value_counts()
```

### 6 - clean ` gender` column 
> tip : you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
candy['Q2: GENDER'] = candy['Q2: GENDER'].fillna('I\'d rather not say')
candy['Q2: GENDER'].value_counts()
```

### 7 - clean ` age` column 
> tip : note the type of age column , and NaN values
 - try to use `to_numeric` 
 - tip : you should choose which best value to put in place of these NaNs ( mean, median, ....etc)

```python
# lengthhh = candy['age'].str.len()
# candy.loc[lengthhh > 2 , 'age'] = np.nan
# candy.loc[lengthhh < 1 , 'age'] = np.nan
# candy['age'] = candy['age'].str.strip('nou?')
candy['age'] = pd.to_numeric(candy['age'], errors='coerce')
age_mean = candy['age'].median()
candy['age']  = candy['age'].astype('int')
candy['age'] = candy['age'].fillna(age_mean)
candy['age'].unique()
```


### 8 - clean ` country` column
> tip : note that country column has country names with different writing ways 
  - use fuzzywuzzy to deal with this problem
  - you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
category = pd.DataFrame({'Count' : pd.Categorical(["united states" , "usa", "canada" , "uk" , "us", "america"])})
candy['country'] = candy['country'].str.lower()
candy['country'] = candy['country'].str.strip()
candy['country'] = candy['country'].str.strip('.\'1234567890')
candy['country'] = candy['country'].replace("." , "")
for count in category['Count'] :
    matches = process.extract(count, candy['country'], limit = candy.shape[0])
    for potential_match in matches:
        if potential_match[1] >= 65:
            candy.loc[candy['country'] == potential_match[0] , 'country'] = count
mapping = {'united states' : 'usa' , 'us' : 'usa' , 'united states of america' : 'usa', 'united kingdom' : 'uk' ,"new york": "usa" ,"california" : 'usa', 'new jersey' : 'usa'}
candy['country']  = candy['country'].replace(mapping)
lengthh = candy['country'].str.len()
candy.loc[lengthh >14, "country"] = np.nan
candy.loc[lengthh < 2, "country"] = np.nan
candy['country'] = candy['country'].fillna('unkown')
candy['country'].unique()
```

### 9 - clean ` area` column
> tip : area column has the same problem as country columns , but looking at all it's unique values may give us another way to deal with it rather than try to fix it

```python
candy['area'] = candy['area'].str.lower()
candy['area'] = candy['area'].str.strip()
candy['area'] = candy['area'].str.strip('1234567890')
lengthh = candy['area'].str.len()
candy.loc[lengthh >20, "area"] = np.nan
candy.loc[lengthh <3, "area"] = np.nan
candy['area'] = candy['area'].fillna('unkown') 
counts = candy['area'].value_counts()
candy['area'] = candy['area'][~candy['area'].isin(counts[counts < 3].index)]
candy['area'].value_counts()
```

### 10 - clean ` Q6` columns
>- tip : Q6 has all kinds of chocolate bars and questions with three distinct options to choose from (Joy, Meh, Despair).
- so all the column that has Q6 is a candy name , so you should go through all these column and you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
f=candy['Q6 | 100 Grand Bar'].mode()
candy['Q6 | 100 Grand Bar'] = candy['Q6 | 100 Grand Bar'].fillna(f)
candy['Q6 | 100 Grand Bar'].value_counts()
```

### 11 - clean ` dress` column
>tip : you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
candy['dress'].unique()
candy['dress'] = candy['dress'].fillna('other')
candy['dress'].value_counts()
```

### 12 - clean ` day` column
>tip : you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
candy['day'] = candy['day'].fillna('do not like both')
candy['day'].value_counts()
```
